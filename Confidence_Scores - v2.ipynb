```python
y_dict={}
a_dict={}
count_am=0
count_yt=0
for id in common_ids:
  yt=str(pd.read_csv('CSV/YouTube/'+id)['Transcript'])
  am=str(pd.read_csv('CSV/Amazon/'+id)['Transcript'])
  #remove punctuation

  yt=yt.split().lower()
  am=am.split().lower()
  for slur in lex:
    if(slur=='you'):
      continue
    if(slur in yt):
      y_dict=update_slurs(yt, y_dict, slur)
      count_yt+=1
    if(slur in am):
      a_dict=update_slurs(am, a_dict, slur)
      count_am+=1
```

```python
def update_slurs(transcript, dictionary, word):
  if(word in dictionary.keys()):
    dictionary[word]+=1
  else:
    dictionary[word]=1
  return dictionary
```

#Slur Analysis (Old)

```python
count=0
ratio=[]
for f in os.listdir("Channels/New Trans/"):
      print(f)
      df=pd.read_csv("Channels/New Trans/"+f)
      count=0
      try:
        for i in df['Transcript']:
          sent_text = nltk.sent_tokenize(i)
          for sentence in sent_text:
            sent_words=sentence.split()
            for word in lex:
              if word in sent_words:
                count+=1
                break
      except:
        print("Error")
      print("Number of transcripts where swear words occur: ", count)
      print("Total number of transcripts:",len(df))
      if(len(df)>0):
        print("Ratio: ",count/len(df))
        ratio.append(count/len(df))
      else:
        ratio.append(0)
      count=0
      print('\n')
```
