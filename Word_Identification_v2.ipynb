```python
from collections import Counter
```

```python
import math

def counter_cosine_similarity(c1, c2):
    terms = set(c1).union(c2)
    dotprod = sum(c1.get(k, 0) * c2.get(k, 0) for k in terms)
    magA = math.sqrt(sum(c1.get(k, 0)**2 for k in terms))
    magB = math.sqrt(sum(c2.get(k, 0)**2 for k in terms))
    return dotprod / (magA * magB)

def length_similarity(c1, c2):
    lenc1 = sum(c1.values())
    lenc2 = sum(c2.values())
    return min(lenc1, lenc2) / float(max(lenc1, lenc2))

def similarity_score(l1, l2):
    c1, c2 = Counter(l1), Counter(l2)
    return length_similarity(c1, c2) * counter_cosine_similarity(c1, c2)
```

```python
def find_matching(seq_am, seq_yt, slur_word):
  #remove punctuation
  seq_am = re.sub(r'[^\w\s]','', seq_am)
  seq_yt = re.sub(r'[^\w\s]','', seq_yt)
  seq_am=seq_am.lower().split()
  seq_yt=seq_yt.lower().split()

  end_index=len(seq_yt)-1
  start_index=0
  for i in range(seq_am.index(slur_word)):
    try:
      if(seq_yt[i] in seq_am):
        start_index=i
    except:
      continue
  ind=seq_am.index(slur_word)+1
  for i in range(ind, len(seq_am)):
    if(seq_am[i] in seq_yt[i:len(seq_yt)]):
      end_index=seq_yt.index(seq_am[i])
      break
  return(' '.join(seq_yt[start_index:end_index+1]))
```

```python
import re
def string_to_list(s):
  s = re.sub(r'[^\w\s]','',s)
  s= s.split()
  return s
final_addr="Final Results/"
final_addr="Preprint Results/"
model_name="XLNet"
transcript_type="YouTube"
p_value=5

csvf=pd.read_csv(final_addr+"P10_"+transcript_type+"_Normalized_AddPluralForms_"+model_name+".csv")
count=0


video_ids_all, aw, sc, ground, pres=[], [], [], [], []
for i in range(len(csvf)):
  actual_words=string_to_list(csvf['Words'][i])
  all_scores=string_to_list(csvf['Scores'][i])

  aw.append(actual_words[:p_value])
  sc.append(all_scores[:p_value])
  ground.append(csvf['Ground Truth'][i])
  video_ids_all.append(csvf['Video Id'][i])
  #print(actual_words)
  #print(str(csvf['Ground Truth'][i]))
  if str(csvf['Ground Truth'][i]) in actual_words[:p_value]:
    count+=1
    pres.append('Present')
  else:
    pres.append('Absent')

print(len(aw))
print(len(sc))
print(len(ground))
print(len(video_ids_all))
print(len(pres))
print(count)

list_data=[]
for i in range(len(aw)):
  list_data.append((video_ids_all[i], ground[i], aw[i], sc[i], pres[i]))

df_to_save = pd.DataFrame(list_data, columns =['Video Id', 'Ground Truth','Words','Scores', 'Present'])
df_to_save.to_csv(final_addr+"P"+str(p_value)+"_"+transcript_type+"_Normalized_AddPluralForms_"+model_name+".csv")
```
